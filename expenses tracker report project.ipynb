```python
import pandas as pd
import matplotlib.pyplot as plt

File = "expenses.csv"
                           
```

```python
def load_data():
    try:
        df = pd.read_csv(FILE)
    except FileNotFoundError:
        columns =  ["date","user_id","monthly_income","monthly_expense_total",
                   "savings_rate","budget_goal","financial_scenario","credit_score",
                   "debt_to_income_ratio","loan_payment","investment_amount",
                   "subscription_services","emergency_fund","transaction_count",
                   "fraud_flag","discretionary_spending","essential_spending",
                   "income_type","rent_or_mortgage","category","cash_flow_status",
                   "financial_advice_score","financial_stress_level",
                   "actual_savings","savings_goal_met"]
        
        df = pd.DataFrame(columns=columns)
        df.to_csv(FILE, index=False)
    return df
    
```

```python
def add_record():
    df = load_data()

    data = {
        "date": input("Enter date (YYYY-MM-DD):"),
        "user_id": input("User ID:"),
        "monthly_income": float(input("Monthly Income:")),
        "monthly_expense_total": float(input("Total Expense:")),
        "savings_rate": float(input("Savings Rate (%):")),
        "budget_goal": float(input("Budget Goal:")),
        "financial_scenario": input("Scenario (stable/risk/etc):"),
        "credit_score": int(input("Credit Score:")),
        "debt_to_income_ratio": float(input("Loan Payment:")),
        "loan_payment": float(input("Loan Payment:")),
        "investment_amount": float(input("Investment Amount:")),
        "subsciption_services": int(input("Subsciptions count:")),
        "emergency_fund": float(input("Emergency Fund:")),
        "transaction_count": int(input("Transaction Count:")),
        "fraud_flag": int(input("Fraud (0/1):")),
        "discretionary_spending": float(input("Extra Spending:")),
        "essential_spending": float(input("Essential Spending:")),
        "income_type": input("Income Type:"),
        "rent_or_mortgage": input("Rent/Mortgage:"),
        "category": input("Category:"),
        "cash_flow_status": input("Cash Flow Status:"),
        "financial_advice_score": float(input("Advice Score:")),
        "financial_stress_level": float(input("Stress Level:")),
        "actual_savings": float(input("Actual Savings:")),
        "savings_goal_met": int(input("Goal Met (0/1):"))
    }

    df = pd.concat([df, pd.DataFrame([data])], ignore_index=True)
    df.to_csv(FILE, index=False)

    print("Record added successfully!\n")
```

```python
#summary analysis
def summary():
    df = load_data()

    print("\n SUMMARY")
    print("Total Users:", df["user_id"].nunique())
    print("Average Income:", df["monthly_income"].mean())
    print("Average Expense:", df["monthly_expense_total"].mean())
    print("Average Savings:", df["actual_savings"].mean())
    print()
```

```python
def plot_income_expense():
    df = load_data()

    df[["monthly_income","monthly_expense_total"]].mean().plot(kind="bar")
    plt.title("Average Income vs Expense")
    plt.xlabel("Type")
    plt.ylabel("Amount")
    plt.show()
```

```python
#Visualization 2: Spending Type
def plot_spending():
    df = load_data()

    data = df[["essential_spending", "discretionary_spending"]].sum()
    data.plot(kind="pie", autopct="%1.1f%%")
    plt.title("Essential vs Discretionary Spending")
    plt.ylabel("")
    plt.show()
```

```python
# Visualization 3: savings goal
def plot_savings_goal():
    df = load_data()

    goal_met = df["savings_goal_met"].value_counts()
    goal_met.plot(kind="bar")
    plt.title("Savings Goal Achievement")
    plt.xlabel("0 = Not Met, 1 = Met")
    plt.ylabel("Count")
    plt.show()
```

```python
#Menu
def menu():
    while True:
        print("------ Expense Tracker ------")
        print("1. Add Record")
        print("2. View Summary")
        print("3. Income vs Expense Graph")
        print("4. Spending Breakdown")
        print("5. Savings Goal Analysis")
        print("6. Exit")

        choice = input("Enter choice:")

        if choice == "1":
            add_record()
        elif choice == "2":
            summary()
        elif choice == "3":
            plot_income_expense()
        elif choice == "4":
            plot_spending()
        elif choice == "5":
            plot_savings_goal()
        elif choice == "6":
            break
        else:
            print("Invalid choice\n")
```
